```python
import pandas as pd, numpy as np
from pathlib import Path
```

```python
root_folderpath = Path(r"C:\Users\jasonjia\Dropbox\shared_spaces\Jason-Kilian\Danish Macro\disaggregated_accounts\process_data_for_tableau")
df1_filepath = Path(root_folderpath / r"data\matrices_individual\Domestic consumer spending matrix (predicted).csv")
df2_filepath = Path(root_folderpath / r"data\matrices_individual\Labor compensation.csv")
```

# Check domestic consumer spending matrix (predicted)

```python
df1 = pd.read_csv(df1_filepath, low_memory=False)
df1
```

```python
# Where do consumers living in KÃ¸benhavn spend?
df1_consumer = df1[df1['cust_muni'] == 101].groupby(['spend_muni']).sum()['s_'].reset_index()
df1_consumer['share'] = np.round(df1_consumer['s_'] / df1_consumer['s_'].sum() * 100, 3)
assert(df1_consumer.iloc[0,2] == 5.417) # Check Copenhagen
assert(df1_consumer.iloc[1,2] == 2.771) # Check Frederiksberg
df1_consumer
```

```python
# Where do the Danish consumers of KÃ¸benhavn producers live?
df1_producer = df1[df1['spend_muni'] == 101].groupby(['cust_muni']).sum()['s_'].reset_index()
df1_producer['share'] = np.round(df1_producer['s_'] / df1_producer['s_'].sum() * 100, 3)
assert(df1_producer.iloc[0,2] == 2.294) # Check Copenhagen
assert(df1_producer.iloc[1,2] == 1.791) # Check Frederiksberg
df1_producer
```

# Check labor compensation matrix

```python
df2 = pd.read_csv(df2_filepath, low_memory=False)
df2
```

```python
# Where does labor compensation paid by producers in KÃ¸benhavn flow to?
df2_producer = df2[df2['prod_muni'] == 101].groupby(['cust_muni']).sum()['comp_'].reset_index()
df2_producer['share'] = np.round(df2_producer['comp_'] / df2_producer['comp_'].sum() * 100, 2)
assert(df2_producer.iloc[0,2] == 43.88) # Check Copenhagen
assert(df2_producer.iloc[1,2] == 7.50) # Check Frederiksberg
df2_producer
```

```python
# Where does labor compensation to consumers in KÃ¸benhavn come from?
df2_consumer = df2[df2['cust_muni'] == 101].groupby(['prod_muni']).sum()['comp_'].reset_index()
df2_consumer['share'] = np.round(df2_consumer['comp_'] / df2_consumer['comp_'].sum() * 100, 2)
assert(df2_consumer.iloc[0,2] == 57.74) # Check Copenhagen
assert(df2_consumer.iloc[1,2] == 4.29) # Check Frederiksberg
df2_consumer
```
